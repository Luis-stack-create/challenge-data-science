### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1=tienda1['Precio'].sum()
ingreso_total_tienda2=tienda2['Precio'].sum()
ingreso_total_tienda3=tienda3['Precio'].sum()
ingreso_total_tienda4=tienda4['Precio'].sum()
print("El ingreso total por cada tienda en orden es:")
print(f"El ingreso Total de la tienda 1 es: ${ingreso_total_tienda1:,.2f}")
print(f"El ingreso Total de la tienda 2 es: ${ingreso_total_tienda2:,.2f}")
print(f"El ingreso Total de la tienda 3 es: ${ingreso_total_tienda3:,.2f}")
print(f"El ingreso Total de la tienda 4 es: ${ingreso_total_tienda4:,.2f}")

```

```python
import matplotlib.pyplot as plt

ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
tiendas = ['Ingreso Tienda 1', 'Ingreso Tienda 2', 'Ingreso Tienda 3', 'Ingreso Tienda 4']

# Crear gráfico de pastel
plt.figure(figsize=(8, 7))
plt.pie(ingresos, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=['#042940','#005C53','#9FC131','#D6D58E'])
plt.title('Ingreso Total por cada Tienda')
plt.axis('equal')
plt.show()
```

# 2. Ventas por categoría

```python
cat1=tienda1['Categoría del Producto'].value_counts().head(3)
```

```python
cat2=tienda2['Categoría del Producto'].value_counts().head(3)
```

```python
cat3=tienda3['Categoría del Producto'].value_counts().head(3)
```

```python
cat4=tienda4['Categoría del Producto'].value_counts().head(3)
```

```python
cat_total=pd.concat([cat1,cat2,cat3,cat4],axis=1)
cat_total.columns=['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
cat_total
```

```python

plt.figure(figsize=(8,7))

# Colores que coinciden con la imagen
colors = ['#042940', '#005C53', '#9FC131']

# Transponer el DataFrame para invertir filas y columnas
ax = cat_total.T.plot(kind='bar',
                   color=colors,
                   edgecolor='black',
                   linewidth=1.2,
                   ax=plt.gca())

plt.title("Lo que más se vende en cada tienda", fontsize=14, fontweight='bold')
plt.ylabel("Cantidad de productos vendidos", fontsize=12)
plt.xlabel("Tiendas", fontsize=12)

# Rotar las etiquetas del eje x
plt.xticks(rotation=45, ha='right')

# Colocar la leyenda fuera del gráfico a la derecha
plt.legend(title="Categorías", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
# 💾 Guardar el gráfico como imagen PNG
plt.tight_layout()
plt.savefig('grafico2_categorias_populares.png', dpi=300, bbox_inches='tight')
# Agregar etiquetas de valores al final de cada barra
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2,       # posición X centrada
            p.get_height() + 0.5,              # un poco por encima de la barra
            f"{int(p.get_height())}",          # valor entero
            ha='center', va='bottom', fontsize=9)

plt.show()
```
