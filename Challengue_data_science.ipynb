### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1=tienda1['Precio'].sum()
ingreso_total_tienda2=tienda2['Precio'].sum()
ingreso_total_tienda3=tienda3['Precio'].sum()
ingreso_total_tienda4=tienda4['Precio'].sum()
print("El ingreso total por cada tienda en orden es:")
print(f"El ingreso Total de la tienda 1 es: ${ingreso_total_tienda1:,.2f}")
print(f"El ingreso Total de la tienda 2 es: ${ingreso_total_tienda2:,.2f}")
print(f"El ingreso Total de la tienda 3 es: ${ingreso_total_tienda3:,.2f}")
print(f"El ingreso Total de la tienda 4 es: ${ingreso_total_tienda4:,.2f}")

```

```python
import matplotlib.pyplot as plt

ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
tiendas = ['Ingreso Tienda 1', 'Ingreso Tienda 2', 'Ingreso Tienda 3', 'Ingreso Tienda 4']

# Crear gráfico de pastel
plt.figure(figsize=(8, 7))
plt.pie(ingresos, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=['#042940','#005C53','#9FC131','#D6D58E'])
plt.title('Ingreso Total por cada Tienda')
plt.axis('equal')
plt.tight_layout()
plt.savefig('grafico1_ingreso_total_por_tienda.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 2. Ventas por categoría

```python
cat1=tienda1['Categoría del Producto'].value_counts().head(3)
```

```python
cat2=tienda2['Categoría del Producto'].value_counts().head(3)
```

```python
cat3=tienda3['Categoría del Producto'].value_counts().head(3)
```

```python
cat4=tienda4['Categoría del Producto'].value_counts().head(3)
```

```python
cat_total=pd.concat([cat1,cat2,cat3,cat4],axis=1)
cat_total.columns=['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
cat_total
```

```python

plt.figure(figsize=(8,7))

# Colores que coinciden con la imagen
colors = ['#042940', '#005C53', '#9FC131']

# Transponer el DataFrame para invertir filas y columnas
fig2 = cat_total.T.plot(kind='bar',
                   color=colors,
                   edgecolor='black',
                   linewidth=1.2,
                   ax=plt.gca())

plt.title("Categoría mas vendida por tienda", fontsize=14, fontweight='bold')
plt.ylabel("Cantidad de productos vendidos", fontsize=12)
plt.xlabel("Tiendas", fontsize=12)

# Rotar las etiquetas del eje x
plt.xticks(rotation=45, ha='right')

# Colocar la leyenda fuera del gráfico a la derecha
plt.legend(title="Categorías", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

# Agregar etiquetas de valores al final de cada barra
for p in fig2.patches:
    fig2.text(p.get_x() + p.get_width()/2,       # posición X centrada
            p.get_height() + 0.5,              # un poco por encima de la barra
            f"{int(p.get_height())}",          # valor entero
            ha='center', va='bottom', fontsize=9)
# 💾 Guardar el gráfico como imagen PNG
plt.tight_layout()
plt.savefig('grafico2_categorias_populares.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
cal1=round(tienda1['Calificación'].mean(),2)
cal2=round(tienda2['Calificación'].mean(),2)
cal3=round(tienda3['Calificación'].mean(),2)
cal4=round(tienda4['Calificación'].mean(),2)
cal_total = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [cal1, cal2, cal3, cal4]
})
print(cal_total)
```

```python
# ✅ Ordenar de mayor a menor calificación
cal_total = cal_total.sort_values(by='Calificación Promedio', ascending=False)

# Crear gráfico de barras horizontales
plt.figure(figsize=(8,6))
barras = plt.barh(cal_total['Tienda'], cal_total['Calificación Promedio'],
                color=['#042940','#005C53','#9FC131','#D6D58E'],
                edgecolor='black')

# Invertir el eje Y para que la mejor tienda aparezca arriba
plt.gca().invert_yaxis()

# Título y etiquetas
plt.title('Calificación Promedio por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Calificación Promedio', fontsize=12)
plt.ylabel('Tiendas', fontsize=12)
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Agregar etiquetas al final de cada barra
for bar in barras:
    width = bar.get_width()
    plt.text(width + 0.05,
             bar.get_y() + bar.get_height()/2,
             f'{width:.2f}', ha='left', va='center', fontsize=10)

# Ajustar márgenes
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos

##Productos mas Vendidos por tienda

```python
productos_mas_vendidos1 = tienda1['Producto'].value_counts().head()
productos_mas_vendidos2=  tienda2['Producto'].value_counts().head()
productos_mas_vendidos3=  tienda3['Producto'].value_counts().head()
productos_mas_vendidos4=  tienda4['Producto'].value_counts().head()
```

```python
prod_mas_total = pd.concat(
    [productos_mas_vendidos1, productos_mas_vendidos2, productos_mas_vendidos3, productos_mas_vendidos4],
    axis=1
)
prod_mas_total.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Reemplazar NaN por 0
prod_mas_total = prod_mas_total.fillna(0)

prod_mas_total
```

```python
import seaborn as sns

# --- Crear el mapa de calor ---
plt.figure(figsize=(8, 7))
sns.heatmap(
    prod_mas_total,
    annot=True,          # muestra los valores
    fmt=".1f",           # formato de número (1 decimal)
    cmap="YlGnBu",       # paleta de color: amarillos → verdes → azules
    linewidths=0.5,      # líneas divisorias suaves
    cbar_kws={'label': 'Porcentaje de ventas (%)'}  # título de la barra lateral
)

plt.title("Mapa de Calor: Productos más Vendidos por Tienda", fontsize=16, fontweight='bold')
plt.xlabel("Tiendas", fontsize=12)
plt.ylabel("Productos", fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##Productos menos vendidos por Tienda

```python
#usamos tail() para ver las ultimas filas
productos_menos_vendidos1 = tienda1['Producto'].value_counts().tail()
productos_menos_vendidos2 = tienda2['Producto'].value_counts().tail()
productos_menos_vendidos3 = tienda3['Producto'].value_counts().tail()
productos_menos_vendidos4 = tienda4['Producto'].value_counts().tail()
```

```python
prod_menos_total = pd.concat(
    [productos_menos_vendidos1, productos_menos_vendidos2,productos_mas_vendidos3,productos_mas_vendidos4],
    axis=1
)
prod_menos_total.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
prod_menos_total = prod_menos_total.fillna(0)
prod_menos_total
```

```python
# --- Crear el mapa de calor ---
plt.figure(figsize=(8, 7))
sns.heatmap(
    prod_menos_total,
    annot=True,          # muestra los valores
    fmt=".1f",           # formato de número (1 decimal)
    cmap="YlGnBu",       # paleta de color: amarillos → verdes → azules
    linewidths=0.5,      # líneas divisorias suaves
    cbar_kws={'label': 'Porcentaje de ventas (%)'}  # título de la barra lateral
)

plt.title("Mapa de Calor: Productos menos Vendidos por Tienda", fontsize=16, fontweight='bold')
plt.xlabel("Tiendas", fontsize=12)
plt.ylabel("Productos", fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
costo_envio1 = round(tienda1['Costo de envío'].mean(),2)
costo_envio2 = round(tienda2['Costo de envío'].mean(),2)
costo_envio3 = round(tienda3['Costo de envío'].mean(),2)
costo_envio4 = round(tienda4['Costo de envío'].mean(),2)
```

```python
costo_envio_total=pd.DataFrame({
    'Tienda':['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],
    'Costo de Envio Promedio':[costo_envio1,costo_envio2,costo_envio3,costo_envio4]
})
print(costo_envio_total)
```

```python
plt.figure(figsize=(8,6))
plt.plot(
    costo_envio_total['Tienda'],
    costo_envio_total['Costo de Envio Promedio'],
    marker='o',
    linewidth=2,
    color='#005C53'
)

plt.title('Tendencia del Costo de Envío Promedio por Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Costo de Envío Promedio (S/.)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Etiquetas de valor en cada punto
for i, valor in enumerate(costo_envio_total['Costo de Envio Promedio']):
    plt.text(i, valor + 0.1, f"S/ {valor:.2f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
